### ✅ Step-by-Step Plan: Compare Predictions with Ground Truth

### ✅ Step 1: Install Required Tools and Define Utilities

```python
import os
import cv2
import matplotlib.pyplot as plt

# Define paths
val_img_dir = "/content/drive/MyDrive/Colab Notebooks/FYP_Model_NEW/datasets/processed/COCO_YOLO_FINAL/val/images/"
gt_label_dir = "/content/drive/MyDrive/Colab Notebooks/FYP_Model_NEW/datasets/processed/COCO_YOLO_FINAL/val/labels/"
pred_label_dir = "runs/detect/predict_val/labels/"  # relative path from notebook

# Load COCO class names (adjust if you're using a subset)
class_names = ['person', 'bicycle', 'car', 'motorcycle', 'bus', 'truck',
               'traffic light', 'stop sign', 'chair', 'bench', 'bottle']

# Colors for GT and prediction
COLOR_GT = (0, 255, 0)      # Green for ground truth
COLOR_PRED = (255, 0, 0)    # Red for prediction
```

### ✅ Step 2: Function to Draw Bounding Boxes

```python
def draw_boxes(image, label_path, color, label_type="GT"):
    if not os.path.exists(label_path):
        print(f"{label_type} label not found: {label_path}")
        return image

    h, w, _ = image.shape
    with open(label_path, 'r') as f:
        for line in f.readlines():
            cls_id, x, y, bw, bh = map(float, line.strip().split())
            cls_id = int(cls_id)
            x1 = int((x - bw / 2) * w)
            y1 = int((y - bh / 2) * h)
            x2 = int((x + bw / 2) * w)
            y2 = int((y + bh / 2) * h)
            label = class_names[cls_id] if cls_id < len(class_names) else str(cls_id)
            cv2.rectangle(image, (x1, y1), (x2, y2), color, 2)
            cv2.putText(image, f"{label} ({label_type})", (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
    return image
```

### ✅ Step 3: Compare Predictions and Ground Truth for Sample Images

```python
# Choose 3 random image filenames (you can customize this list)
sample_filenames = ['000000000064.jpg', '000000000071.jpg', '000000000151.jpg']

for fname in sample_filenames:
    img_path = os.path.join(val_img_dir, fname)
    gt_label_path = os.path.join(gt_label_dir, fname.replace('.jpg', '.txt'))
    pred_label_path = os.path.join(pred_label_dir, fname.replace('.jpg', '.txt'))

    # Load image
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Copy image for drawing
    img_gt = img.copy()
    img_pred = img.copy()

    # Draw boxes
    img_gt = draw_boxes(img_gt, gt_label_path, COLOR_GT, "GT")
    img_pred = draw_boxes(img_pred, pred_label_path, COLOR_PRED, "PRED")

    # Plot side-by-side
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.title("Ground Truth")
    plt.imshow(img_gt)
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Model Prediction")
    plt.imshow(img_pred)
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

🔁 Step 1: Convert .avi to .mp4 using OpenCV and FFmpeg

```python
import cv2

# Define paths
avi_path = "/content/runs/detect/predict2/head-pose-face-detection-female.avi"
mp4_path = "/content/runs/detect/predict2/head-pose-face-detection-female.mp4"

# Open .avi video
video = cv2.VideoCapture(avi_path)

# Get video details
fps = video.get(cv2.CAP_PROP_FPS)
width  = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define VideoWriter for .mp4 output
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec
out = cv2.VideoWriter(mp4_path, fourcc, fps, (width, height))

while video.isOpened():
    ret, frame = video.read()
    if not ret:
        break
    out.write(frame)

video.release()
out.release()

print("✅ Converted to MP4:", mp4_path)
```

👁️ Step 2: Display the Converted .mp4 Inline

```python
from IPython.display import HTML
from base64 import b64encode

def show_video(video_path, width=640, height=480):
    mp4 = open(video_path, 'rb').read()
    data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
    return HTML(f"""
        <video width="{width}" height="{height}" controls>
            <source src="{data_url}" type="video/mp4">
        </video>
    """)

# Show the converted video
show_video("/content/runs/detect/predict2/head-pose-face-detection-female.mp4")
```
